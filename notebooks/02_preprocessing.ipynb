```python
import pandas as pd
import numpy as np

data = pd.read_csv("E:/Programing/CyberSentinel/data/CICIDS2017/merged.csv", low_memory=False)
print("Loaded merged CSV:", data.shape)
```

```python
# Replace infinite values with NaN
data.replace([np.inf, -np.inf], np.nan, inplace=True)

# Drop rows with missing values
data.dropna(inplace=True)

print("After removing NaN:", data.shape)
```

```python
data.rename(columns=lambda x: x.strip(), inplace=True)
print("Removed leading and trailing whitespace from the column names.")
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
data["Label"] = le.fit_transform(data["Label"])

print("Classes:", le.classes_)
```

```python
X = data.drop("Label", axis=1)
y = data["Label"]

print("Features shape:", X.shape)
print("Labels shape:", y.shape)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

print("X_scaled shape:", X_scaled.shape)
```
